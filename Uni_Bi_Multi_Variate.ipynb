```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\admin\Downloads\Iris.csv',index_col=0)
df.head()
```

```python
# remove Iris from the Species
df['Species'] = df['Species'].str.split('-',1).str[1]
df
```

# Univariate

```python
x = df['Species']
y = df['SepalLengthCm']

sns.relplot(data=df,x=x,y=y,hue=df['Species'])
plt.show()
```

# BiVariate

```python
b = df['SepalWidthCm']
a = df['SepalLengthCm']

sns.relplot(data=df,x=a,y=b,hue="Species")
plt.show()
```

# Multivariate

```python
sns.pairplot(data=df, hue="Species")
```

```python
sns.heatmap(df.corr(),linewidths=1,annot=True)
```
